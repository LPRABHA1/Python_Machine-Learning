```python
def img_list():
    img_list1 = []
    for indx in range(0,50,1):
        img = z_train[indx]
        img2 = np.zeros((32,32,3), dtype = np.int )
        i = 0
        j = 0
        l = 0

        while l < 32:
            j = 0
            while j < 32:
                k = 0
                img2[l][j][k] = img[i]
                img2[l][j][k+1] = img[i+1024]
                img2[l][j][k+2] = img[i+2048]
                i = i + 1
                j = j + 1
            l = l + 1
        img_list1.append(img2)
        
#        plt.figure(figsize=(0.5,0.5))
        # shows the image 
#        plt.imshow(img2) 
#        plt.show() 
        
    return(img_list1)
```

```python
img_list1 = img_list()
```

```python
img_arr = np.array(z_train)
```

```python
label_arr = np.array(label)
```

```python
img_arr.shape
```

```python
## Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(z_train,label_arr)
print(model)
```

```python
import sklearn.metrics as metrics
```

```python
# make predictions
expected = label_arr
predicted = model.predict(z_train)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
# Knn- classification
# k-Nearest Neighbor Classification
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(z_train,label_arr)
print(model)
```

```python
# make predictions
expected = label_arr
predicted = model.predict(z_train)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(z_train,label_arr)
print(model)
```

```python
# make predictions
expected = label_arr
predicted = model.predict(z_train)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
print(cm)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver = 'lbfgs',random_state=10)
model.fit(z_train,label_arr)
print(model)
```

```python
# make predictions
expected = label_arr
predicted = model.predict(z_train)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
print(cm)
```
